# Dataset Distribution

```python
import numpy as np
import matplotlib.pyplot as plt
import torch
from torchvision import datasets, transforms

# 加载CIFAR-10数据集
transform = transforms.Compose([transforms.ToTensor(), transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5)),])
train_dataset = datasets.ImageFolder(root='/workspace/data/splited_dataset/train', transform=transform)

train_img = [train_dataset[i][0] for i in range(len(train_dataset))]
train_label = [train_dataset[i][1] for i in range(len(train_dataset))]

print(len(train_img), len(train_label))
```

```python

# 1. 类别分布
temp = ['cold', 'cool', 'hot', 'normal', 'warm']

unique_classes, class_counts = np.unique(train_label, return_counts=True)
colors = plt.cm.get_cmap('tab10', len(unique_classes))
bars = plt.bar(unique_classes, class_counts, color=colors(np.arange(len(unique_classes))))
plt.xticks(unique_classes, [str(cls) for cls in temp])
plt.xlabel('Class')
plt.ylabel('Number of Samples')
plt.title('5 Class Train Dataset Distribution')

# legend_labels = ['Class ' + str(cls) for cls in temp]
# plt.legend(bars, legend_labels, loc='upper right')

plt.show()

# 2. 图像分布
image_shapes = [str(img.shape) for img in train_img]
unique_shapes, shape_counts = np.unique(image_shapes, return_counts=True)
plt.bar(unique_shapes, shape_counts)
# plt.xticks(rotation=45)
plt.xlabel('Image Shape')
plt.ylabel('Number of Samples')
plt.title('CIFAR-10 Image Shape Distribution')
plt.show()
```

# Test Dataset Distribution

```python
from cgi import test
import numpy as np
import matplotlib.pyplot as plt
import torch
from torchvision import datasets, transforms

# 加载CIFAR-10数据集
transform = transforms.Compose([transforms.ToTensor(), transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5)),])
test_dataset = datasets.ImageFolder(root='/workspace/data/splited_dataset/test', transform=transform)

test_img = [test_dataset[i][0] for i in range(len(test_dataset))]
test_label = [test_dataset[i][1] for i in range(len(test_dataset))]

print(len(test_img), len(test_label))
```

```python

# 1. 类别分布
temp = ['cold', 'cool', 'hot', 'normal', 'warm']

unique_classes, class_counts = np.unique(test_label, return_counts=True)
colors = plt.cm.get_cmap('tab10', len(unique_classes))
bars = plt.bar(unique_classes, class_counts, color=colors(np.arange(len(unique_classes))))
plt.xticks(unique_classes, [str(cls) for cls in temp])
plt.xlabel('Class')
plt.ylabel('Number of Samples')
plt.title('5 Class Val Dataset Distribution')

# 在每个柱形上显示数字
for bar, count in zip(bars, class_counts):
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 100, int(count), ha='center', va='bottom')

plt.show()

# 2. 图像分布
image_shapes = [str(img.shape) for img in test_img]
unique_shapes, shape_counts = np.unique(image_shapes, return_counts=True)
plt.bar(unique_shapes, shape_counts)
# plt.xticks(rotation=45)
plt.xlabel('Image Shape')
plt.ylabel('Number of Samples')
plt.title('CIFAR-10 Image Shape Distribution')
plt.show()
```
